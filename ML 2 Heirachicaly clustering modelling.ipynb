```python
#Lets preprocess the data

 

from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
```

```python
car_scaled=scaling.fit_transform(cars_cleaned)
```

```python
car_scaled=pd.DataFrame(car_scaled,columns=cars_cleaned.columns)
```

```python
car_scaled
```

```python
#building heirachical model
```

```python


import scipy.cluster.hierarchy as sch



dendogram=sch.dendrogram(sch.linkage(car_scaled,method="ward"))
```

```python


from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
```

```python
clusters=hc.fit_predict(car_scaled)
clusters
```

```python
cars['cluster']=clusters
```

```python
cars
```

```python
#cluster analysis
cars.groupby('cluster').mean()
```

```python
#visualize cluster on base of type & mileage

sns.scatterplot(x=cars['type'],y=cars['mpg'],hue=cars['cluster'])
```

```python
#lets visualise the cluster based on sales and mileage

 

sns.scatterplot(x=cars["sales"],y=cars["mpg"],hue=cars["cluster"])
```
